### LINE CHART


A line chart, also known as a line plot or line graph, is a type of graph that displays data as points connected by straight lines. It is commonly used to represent trends and relationships between continuous variables over a continuous interval or time period

```python
# creating a series data
x = [2001,2002,2003,2004,2005,2006,2007,2008,2009,2010]
years = np.array(x)
print(years)
sales = np.random.randint(50000,100000,10)
print(sales)
```

```python
# plotting a line chart for sales in every year 
plt.plot(years,sales)
# add labels and title
plt.title('Sales per year')
plt.xlabel('Years')
plt.ylabel('Sales')
# used to customize the tick positions and labels on the x-axis and y-axis
plt.xticks(years)
# to display the figure
plt.show()
```

Parameter in Line Chart
* x,y: hese are the input data arrays. x represents the values for the x-axis, and y represents the values for the y-axis.
* Color: Specifies the color of the line. It can be specified as a string (e.g., 'red', 'blue', 'green')
* linestyle(or ls): Specifies the style of the line. It can be '-' for solid, '--' for dashed, ':' for dotted, or '-.' for dash-dot.
* linewidth(or lw): specifies the width of the line.
* marker: Specifies the marker style for data points. It can be 'o' for circles, 's' for squares, '+' for plus signs, etc.
* markersize(ms): Specifies the size of the markers.

```python
plt.plot(years,sales,color='red',ls='--',lw=3,marker='o',ms=5)
plt.title('Sales per year')
plt.xlabel('Years')
plt.ylabel('Sales')
plt.show()
```

#### SCATTER PLOT

A Scatter plot is made between two number columns and are usually drawn to see the trend or relation among the number columns. Below are examples of scatter plots illustrating different types of relationships between variables:

1. Positive Relationship:
   * In a positive relationship, as one variable increases, the other variable also tends to increase.
   * Example: The relationship between hours studied and exam scores.

```python
hours_studied = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [50, 60, 70, 80, 85, 90, 95, 100]
```

```python
# create a scatter plot 
plt.scatter(hours_studied,exam_scores)
plt.title('Positive Relationship: Hours Studied vs Exam Scores')
plt.xlabel('Hours Studied')
plt.ylabel('Exam Scores')
plt.show()
```

2. Negative Relationship:
   * In a negative relationship, as one variable increases, the other variable tends to decrease.
   * Example: The relationship between temperature and sales of hot chocolate.

```python
temperature = [30, 25, 20, 15, 10, 5, 0]
hot_chocolate_sales = [10, 15, 20, 25, 30, 35, 40]
```

```python
plt.scatter(temperature, hot_chocolate_sales)
plt.title('Negative Relationship: Temperature vs Hot Chocolate Sales')
plt.xlabel('Temperature (°C)')
plt.ylabel('Hot Chocolate Sales')
plt.show()
```

3. No Relationship:
   * In a scatter plot with no relationship, there is no clear pattern or correlation between the variables.
   * Example: The relationship between shoe size and IQ scores.

```python
shoe_size = [random.uniform(5, 15) for _ in range(50)]
iq_scores = [random.uniform(70, 130) for _ in range(50)]
```

```python
plt.scatter(shoe_size, iq_scores)
plt.title('No Relationship: Shoe Size vs IQ Scores')
plt.xlabel('Shoe Size')
plt.ylabel('IQ Scores')
plt.show()
```

Parameter in Scatter Plot:
* x, y: Arrays or sequences of data representing the x and y coordinates of the data points.
* s: Size of markers. You can specify a scalar value or an array-like object to define the size of each marker individually.
* c: Color of markers. You can specify a color using a single color format string, an array-like object of the same length as x and y, or a sequence of colors.
* marker: Marker style. Specifies the shape of markers, such as 'o' for circles, 's' for squares, '+' for plus signs, etc.
* alpha: Transparency of markers. It ranges from 0 (completely transparent) to 1 (completely opaque).
* linewidths (or lw): Width of marker edges.
* edgecolors: Color of marker edges.
* cmap: Colormap. If c is an array-like object, cmap is used to map numeric values to colors.
* norm: Normalization for colormap.
* vmin, vmax: Minimum and maximum values for colormap normalization.
* label: Label for the scatter plot, used in legends.legends.

```python
data = sns.load_dataset('iris')
data.head(4)
```

```python
# Make a scatter plot between sepal length and sepal width
plt.scatter(data['sepal_length'],data['sepal_width'],c='tomato',s=20,alpha=1,marker='x')
plt.title('Iris')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width')
plt.show()
```

By looking at the plot, we can conclude that the points are randomly scatter and there is no relation between the sepal_length and sepal_width.

```python
# Make a scatter plot between sepal lenght sepal width, sepal length and petal length
plt.scatter(data['sepal_length'],data['sepal_width'],c='tomato',s=20,alpha=1,marker='x',label='SW')
plt.scatter(data['sepal_length'],data['petal_length'],c='darkgreen',marker='*',label='PL')
plt.title('Iris')
plt.xlabel('sepal_length')
plt.ylabel('sepal_width/petal_length')
plt.legend()
plt.show()
```

```python
# Sub plot : Subplots allow you to create multiple plots within a single figure
# width = 7, height = 2.
plt.figure(figsize=(10,4))
plt.suptitle('Iris')

plt.subplot(1,2,1)
plt.scatter(data['sepal_length'],data['sepal_width'],c='tomato',s=20,alpha=1,marker='x',label='SW')
plt.xlabel('Sepal_length')
plt.ylabel('Sepal_width')

plt.subplot(1,2,2)
plt.scatter(data['sepal_length'],data['petal_length'],c='darkgreen',marker='*',label='PL')
plt.xlabel('Sepal_length')
plt.ylabel('Petal_length')

plt.tight_layout()
plt.show()
```

#### BAR PLOT

Bar plot is made when you have number against each category. It shows bars and the height of bar is used to compare the numbers.

Parameter in barplot:
* x: Array-like or scalar sequence representing the x coordinates of the bars.
* height: Array-like or scalar sequence representing the height of the bars.
* width: Scalar or array-like, optional. Width of the bars. If not specified, a default width will be used, which is 0.8.
* bottom: Scalar or array-like, optional. The y coordinate(s) of the bars' bases. If not specified, defaults to 0.
* align: {'center', 'edge'}, optional. Alignment of the bars with respect to their x coordinates. Default is 'center'.
* color: Color or array-like of colors, optional. Color(s) of the bars.
* edgecolor: Color or array-like of colors, optional. Color(s) of the bar edges.
* linewidth: float or array-like, optional. Width of the bar edges.
* tick_label: Array-like, optional. Labels for the x-axis ticks.
* label: String or array-like, optional. Label(s) for the legend.
* alpha: float, optional. Transparency of the bars, ranging from 0 (completely transparent) to 1 (completely opaque).(completely opaque).

```python
amt = [2500,3700,3000,4200,2200]
cust = ['Ajay','Suresh','Kamla','Neha','Vijay']
```

```python
# Make a vertical barplot b/w amt nad cust
plt.bar(x=cust,height=amt,width=0.8,color=['red','blue','green','magenta','cyan'])
for x,y in zip(range(5),amt):
    plt.text(x-0.2,y+100,str(y))
plt.yticks(range(0,5001,1000))
plt.grid()
plt.show()
```

The barh() function in Matplotlib is used to create horizontal bar plots. It accepts several parameters to customize the appearance of the bar plot. Here are some commonly used parameters:
* y: The y-coordinates of the bars.
* width: The width of the bars. It can be a scalar or an array-like object specifying the width of each bar individually.
* height: The height of the bars. This parameter is typically used in vertical bar plots (bar() function), but it's available in barh() as well.
* left: The x-coordinates of the left edges of the bars.
* color: The color of the bars. It can be a single color string or an array-like object specifying the color of each bar individually.
* align: Specifies the alignment of the bars relative to the y-coordinates. It can be 'center' (default), 'edge', or 'zero'.
* edgecolor: The color of the edges of the bars.
* linewidth: The width of the edges of the bars.
* tick_label: Labels for the bars. It can be a list of strings specifying the labels for each bar.

```python
# Make a horizonatl bar plot b/w amount and cust
plt.barh(y=cust,width=amt,color='darkgreen')
plt.grid()
plt.show()
```

#### GROUPED BAR PLOT

To create a grouped bar plot in Matplotlib, you typically need to specify the positions of the bars for each group, along with the width of the bars. Here's how you can create a grouped bar plot:

```python
python = np.random.randint(30,50,5)
npv=np.random.randint(30,50,5)
```

```python
x = np.arange(1,6)
```

```python
# Make a grouped bar plot between python npv and x
bar_width = 0.3
plt.bar(x=x-bar_width/2,height=python,width=bar_width,label='ITP')
plt.bar(x=x+bar_width/2,height=npv,width=bar_width,label='NPV')
plt.xticks(x,cust)
plt.legend()
plt.show()
```

The position of the bars for the second group is shifted by bar_width/2 to the right to avoid overlapping.

#### STACKED BAR PLOT 

To create a stacked bar plot in Matplotlib, you can use the bar() function multiple times, each time specifying the positions of the bars and their heights. Here's how you can create a stacked bar plot:

```python
# Make a stacked bar plot between  python npv and x
plt.bar(x=cust,height=python,label='ITP')
plt.bar(x=cust,height=npv,bottom=python,label='NPV')
plt.legend()
plt.show()
```

The first call to plt.bar() creates the first set of bars (python), and the second call creates the second set of bars (npv). By specifying bottom=python, the bars for the second set are stacked on top of the bars for the first set.


#### PIE CHART

A Pie Chart is a circular statistical plot that can display only one series of data. The area of the chart is the total percentage of the given data.

Parameter in Pie plot:
* x: The data to be plotted as a pie chart. This can be an array-like object containing numerical values representing the sizes of the wedges.
* labels: Labels for the wedges. This parameter is used to specify the text to be displayed for each wedge of the pie chart.
* explode: Array-like object of the same length as x, specifying the fraction of the radius with which to offset each wedge. This parameter is useful for emphasizing certain wedges by "exploding" them outwards.
* colors: Colors for the wedges. This parameter is used to specify the colors of the wedges. It can be a list of color names or a list of RGB tuples.
* autopct: Format string for the percentage labels. This parameter specifies the format of the percentage labels displayed on each wedge. For example, '%.1f%%' will display one decimal place.
* startangle: The angle by which the start of the pie is rotated counterclockwise from the x-axis. This parameter is useful for changing the starting position of the wedges.
* shadow: Boolean value indicating whether to draw a shadow beneath the pie. Setting shadow=True adds a shadow effect to the pie chart.
* radius: This parameter specifies the radius of the pie chart. By default, the radius is 1, which means the pie chart fits snugly within the axis boundaries. You can adjust this value to change the size of the pie chart.
* labeldistance: The parameter labeldistance controls the radial distance of the labels from the center of the pie chart. It is specified as a fraction of the radius. For example, labeldistance=0.5 places the labels halfway between the center and the edge of the pie chart.
* pctdistance: This parameter controls the radial distance of the autopct labels (percentage labels) from the center of the pie chart. It is specified as a fraction of the radius. Similar to labeldistance, you can adjust this parameter to change the position of the percentage labels.
* counterclock: This parameter specifies the direction in which the pie chart is drawn. By default, counterclock=True, which means the angles are measured counterclockwise. If set to False, the angles will be measured clockwise.

```python
countries = ['USA','France','Germany','India','Russia']
count=[63.76,34.54,45.11,57.43,60.54]
```

```python
# Make a pie plot
plt.pie(x=count,labels=countries,radius=1.3,colors=['green','red','blue','magenta','cyan'],autopct='%0.f%%',
       pctdistance=0.8,shadow=True,labeldistance=1.2,startangle=90,counterclock=False,explode=[0.2,0,0,0,0])
plt.show()
```

```python
# Make a doughnut Chart 
# creates a pie chart with specific parameters and then overlays it with a smaller pie chart to create a doughnut effect. 
plt.pie(x=count,labels=countries,radius=1.3,colors=['green','red','blue','magenta','cyan'],autopct='%0.f%%',
       pctdistance=0.8,shadow=True,labeldistance=1.2,startangle=90,counterclock=False,explode=[0.2,0,0,0,0])
plt.pie(x=[1],colors=['w'],radius=0.8)
plt.show()
```

### HEAT MAP


A heatmap is a graphical representation of data in a matrix format where values are represented as colors. It is particularly useful for visualizing the correlation between variables in a dataset or displaying the magnitude of values across two dimensions.

```python
d=pd.crosstab(index=data_['pclass'],columns=data_['sex'],values=data_['age'],aggfunc='mean')
print(d)
```

```python
# heatmap of the crosstab data
sns.heatmap(d,annot=True,vmin=15,vmax=50,cmap='RdBu')
plt.show()
```
